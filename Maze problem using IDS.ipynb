```python
last_index = 0
visited_node = []

def gen_successors(node):
    """Generate successor nodes

    Args:
      node (tuple): the parent node for generating successors.
    
    Rutrns
      list of tuples: the list of all children nodes that ware generated.

    """
    global last_index
    graph = {
    'A':'S',
    'B':'SDC',
    'C':'JB',
    'D':'SGB',
    'E':'SG',
    'F':'HG',
    'G':'JHFED',
    'H':'IGF',
    'I':'JH',
    'J':'IGC',
    'S':'EDBA',
    }
    
    node_id = node[1]
    level = node[3]
    ret = []
    for c in graph[node[0]]:
        # node (lable, node_id, parent_node_id, level)
        ret += [(c,last_index+1,node[1],node[3]+1)]
        last_index += 1
    return ret

def insert_all(node,fringe):
    """Insert the successor nodes to the fringe.

    Args:
      node (tuple): the parent node for generating successors.
      fringe(list of tuples): the set of all nodes at the end of all visited paths.

    """
    children = gen_successors(node)
    for child in children:
        fringe[0:0] = [child] ##Inset chidren
    print(fringe) #Print 


def is_goal(node):
    """Check whether the node is goal or not.

    Args:
      node (tuple): the tuple node includes '(lable, node_id, parent_node_id, level)'.

    Ruturns:
      bool

    """
    return node[0] == 'I'


def show_result(g):
    """Print the path from start to exit when the exit was found.

    Args:
      g (tuple): the exit node includes (lable, node_id, parent_id, level).
    
    """
    global visited_node
    current_node = g
    parent_index = current_node[1]
    result = []
    while True:
        # print(current_node[0],end=' ')
        result[0:0] = current_node[0]
        parent_index = current_node[2]
        if parent_index == -1:
            break
        current_node = visited_node[parent_index]
    print(" ")
    print(*result, sep=" -> ")


def dfs(start_node,limited_level):
    """ Depth First Search to limited level

    Args:
      start_node (tuple):  the tuple node includes '(lable, node_id, parent_node_id, level)'.
      limited_level (int): the limit level of DFS.
    
    Returns:
      bool: return True if the exit was found.

    """
    global last_index
    global visited_node
    last_index = 0
    fringe = [start_node]
    print(fringe)
    visited_node = {}
    while True:
        if len(fringe) == 0:
            print('Not Found')
            break
        front = fringe[0]
        visited_node[front[1]] = front
        fringe = fringe[1:]
        if is_goal(front):
            show_result(front)
            return True
        if front[3] == limited_level:
            continue
        insert_all(front,fringe)



for i in range(100):
    print('Limit at level '+str(i))
    # node (lable, node_id, parent_node_id, level)
    if dfs(('S',0,-1,0),i):
        break
```
